```python
df_left = pd.DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
                        'A': [0.8, 0.5, 0.2, 0.7],
                        'B': [0.5, 0.2, 0.6, 0.4]})
```

```python
df_right = pd.DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
                        'C': [0.8, 0.6, 0.2, 0.5],
                        'D': [0.1, 0.8, 0.6, 0.5]})
```

```python
df = pd.merge(df_left, df_right, how='outer', on='KEY')
```

```python
df
```

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum(1)
```

```python
df.fillna(0)
```

```python
df.fillna('missing')
```

```python
mean_val = df.iloc[:, 1:5].mean().mean()  # 전체 평균
df.fillna(mean_val)
```

```python
df.fillna(df.mean(numeric_only=True)) # 각 열 평균
```
